## Popularity Based

```python
rating_merge=df1.merge(df,on='ISBN')
```

```python
rating_merge
```

```python
num_rating=rating_merge.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating.rename(columns={'Book-Rating':'num_Rating'},inplace=True)
num_rating
```

```python
avg_rating=rating_merge.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_Rating'},inplace=True)
avg_rating
```

```python
popular_df=num_rating.merge(avg_rating,on='Book-Title')
```

```python
popular_df
```

```python
popular_df['num_Rating'].value_counts()
```

```python
popular_df=popular_df[popular_df['num_Rating']>=250].sort_values('avg_Rating',ascending=False).head(50)
```

```python
popular_df
```

```python
popular=popular_df.merge(df,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Publisher','Image-URL-M','num_Rating','avg_Rating']]
```

## Collabrative Based Recommendation system

```python
x=rating_merge.groupby('User-ID').count()['Book-Rating'] > 250
users=x[x].index
```

```python
users
```

```python
filtered_rating=rating_merge[rating_merge['User-ID'].isin(users)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=50
```

```python
famous_books=y[y].index
```

```python
final=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar_score=cosine_similarity(pt)
```

```python
similar_score.shape
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar=sorted(list(enumerate(similar_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar:
        item=[]
        temp_df=filtered_rating[filtered_rating['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
